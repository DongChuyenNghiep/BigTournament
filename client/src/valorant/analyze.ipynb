### Lấy data

```python
import requests
import pandas as pd
```

```python
dfs = []
matchids = ["3da83392-be3f-4456-a16a-064043f1fb04","c78a04f8-c04b-48da-9ffd-e170399f328b",
            "b66e4235-0fce-40c7-ae31-9b92d226d9f7","11034875-dab1-4dd9-ab7c-9b8ecb4b7a5d",
            "50f1aeab-6490-4723-99de-199dac459c5c","42b13fb8-eb11-49cb-aec9-8421d811407f",
            "9ba36f5a-db1c-47ef-9e55-dcc6ec2a3d26"
]
```

```python
for matchid in matchids:
    api_url = f"https://bigtournament-hq9n.onrender.com/api/valorant/match/{matchid}"
    response = requests.get(api_url)
    if response.status_code == 200:
        data = response.json()
        df = pd.json_normalize(data)
        df["matchid"] = matchid
        dfs.append(df)
        print(f"Lấy data từ {matchid} thành công")
    else:
        print(f"Match Id {matchid} không tồn tại trên hệ thống")
```

```python
if dfs:
    riot_df = pd.concat(dfs,ignore_index=True)
else:
    print("Không có dữ liệu hợp lệ")
```

```python
riot_df
```

```python
column_to_remove = ["coaches","matchInfo.matchId","matchInfo.gameVersion",
                    "matchInfo.region","matchInfo.provisioningFlowId",
                    "matchInfo.customGameName","matchInfo.queueId",
                    "matchInfo.gameMode","matchInfo.isRanked","matchInfo.seasonId"]
```

```python
riot_df.drop(column_to_remove,errors="ignore",inplace=True)
```

```python
riot_df.fillna(0, inplace=True)
```

```python
all_players = []
stats = []
for eachmatch in dfs:
    for player in eachmatch.players:
        for playerstat in player: 
            all_players.append(playerstat)
```

```python
df2 = pd.DataFrame(all_players)
if "stats" in df2.columns:
    # Bung dữ liệu trong cột "stats" thành các cột riêng biệt
    stats_df = pd.json_normalize(df2["stats"])
    
    # Gộp DataFrame chính với DataFrame stats đã bung
    df2 = df2.drop(columns=["stats"]).join(stats_df)

df2
```

```python
drop_col = ["teamId","partyId","competitiveTier","isObserver","playerCard","playerTitle","accountLevel"]
```

```python
df2.drop(columns=drop_col,errors="ignore",inplace=True)
```

```python
df2.fillna(0, inplace=True)
df2
```

```python
df_grouped = df2.groupby(["gameName", "tagLine"], as_index=False).sum(numeric_only=True)
df_grouped
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
numeric_df = df_grouped.select_dtypes(include=["number"])  # Loại bỏ các cột chứa chuỗi

# Tính ma trận tương quan
correlation_matrix = numeric_df.corr()

# Vẽ heatmap của bảng tương quan
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Thêm tiêu đề
plt.title("Bảng đồ tương quan giữa các chỉ số thống kê")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import scipy.cluster.hierarchy as sch
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Giả sử df_grouped đã tồn tại và chứa dữ liệu
# Sắp xếp df_grouped theo số kill giảm dần
df_grouped = df_grouped.sort_values(by="kills", ascending=False)

# Chọn các cột số
numeric_df = df_grouped.select_dtypes(include=["number"])

# Chuẩn hóa dữ liệu để tránh ảnh hưởng bởi đơn vị đo khác nhau
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_df)

# Áp dụng thuật toán KMeans để phân cụm
kmeans = KMeans(n_clusters=5, random_state=42)
df_grouped["Cluster"] = kmeans.fit_predict(scaled_data)

# Vẽ biểu đồ scatter plot với phân cụm
graph = sns.scatterplot(data=df_grouped, x="kills", y="deaths", hue="Cluster", palette="Set1")
graph.set_title("Phân Cụm Người Chơi Dựa Trên Kills & Deaths")
plt.xlabel("Kills")
plt.ylabel("Deaths")
plt.legend(title="Cluster")
plt.show()
```
